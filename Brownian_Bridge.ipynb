```
# -*- coding: utf-8 -*-
"""
Created on Sun Jun 30 20:01:46 2019

@author: Martin Musya
"""

import numpy as np
import matplotlib.pyplot as plt



#from matplotlib import pyplot as plt

#pip install numpy as np
#pip.install matplotlib.pyplot as plt

y0 = 0  #initial stock price
K = 1  #strike price
b = 1  
sigma = 0.50  #volatility in market
r=.05
T = 1  #time in years
N = 252  #number of steps within each simulation
deltat = T / N  #time step
nsim = 5  #number of simulations
discount_factor = np.exp(-r * T)  #discount factor

y = np.zeros([nsim, N])
t = range(0, N, 1)

for j in range(0, nsim - 1):
    y[j, 0] = y0
    xi = np.zeros(N-1)
    for i in range(0, N - 1):
      xi[i] = sigma * np.sqrt(deltat) * np.random.normal(0, 1)
      #y[j, i + 1] = y[j,i]+((b-y[j,i])/(1-i))*deltat+xi[i]
      y[j, i + 1] = y[j,i]*(1-deltat/(1-(i/252)))+xi[i]
    plt.plot(t, y[j])

plt.title(
    'Brownian Bridge \n Sims %d | Steps %d | Sigma %.2f | r %.2f | y0 %.2f' % (nsim, N, sigma, r, y0))
plt.xlabel('Steps')
plt.ylabel('Option Price')
#plt.ylim(60,120)
plt.show()

#plt.plot([1,2,3,4,5,6,7,8,9])
#plt.show()
plt.savefig('Brownian_Bridge.png')


```
