```
# -*- coding: utf-8 -*-
"""
Created on Sun Jun 30 20:01:46 2019

@author: Martin Musya
"""

import numpy as np
import matplotlib.pyplot as plt



#from matplotlib import pyplot as plt

#pip install numpy as np
#pip.install matplotlib.pyplot as plt

r0 = 0  #initial int rate
#b = 24.220  #initial int rate
#a = 0.258
#b = 0.281

a = 0.1146
b = -0.0196
gamma = 0.5


#K = 100  #strike price
#alpha = 0.0866  #risk-free interest rate
#sigma = 0.0004  #volatility in market
sigma = 0.0658  #volatility in market
T = 1  #time in years
N = 252  #number of steps within each simulation
deltat = T / N  #time step
nsim = 5  #number of simulations
discount_factor = np.exp(-r0 * T)  #discount factor
beta = -0.1779  #beta

r = np.zeros([nsim, N])
t = range(0, N, 1)

for y in range(0, nsim - 1):
    r[y, 0] = r0
    for x in range(0, N - 1):
        r[y, x + 1] =r[y, x] + a*(b-r[y, x])*deltat  + sigma * \
                          np.sqrt(abs(r[y, x])) * np.sqrt(deltat) * \
                          np.random.normal(0, 1)
    plt.plot(t, r[y])

plt.title(
    'CIR SR \n Sims %d | Steps %d | Sigma %.2f | a = %.2f | b = %.2f | r0 %.2f' % (nsim, N, sigma, a, b, r0))
plt.xlabel('Steps')
plt.ylabel('Interest rate')
#plt.ylim(60,120)
plt.show()

#plt.plot([1,2,3,4,5,6,7,8,9])
#plt.show()
plt.savefig('CIR_SR.png')

```
