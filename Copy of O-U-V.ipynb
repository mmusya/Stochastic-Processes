```
# -*- coding: utf-8 -*-
"""
Created on Sun Jun 30 20:01:46 2019

@author: Martin Musya
"""

import numpy as np
import matplotlib.pyplot as plt



#from matplotlib import pyplot as plt

#pip install numpy as np
#pip.install matplotlib.pyplot as plt

r0 = 0.06  #initial int rate
K = 100  #strike price
alpha = 0.0866  #risk-free interest rate
#sigma = 0.0004  #volatility in market
sigma = 0.04  #volatility in market
T = 1  #time in years
N = 252  #number of steps within each simulation
deltat = T / N  #time step
nsim = 5  #number of simulations
discount_factor = np.exp(-r0 * T)  #discount factor
beta = -0.1779  #beta

r = np.zeros([nsim, N])
t = range(0, N, 1)

for y in range(0, nsim - 1):
    r[y, 0] = r0
    for x in range(0, N - 1):
        r[y, x +
          1] = r[y, x] + (alpha +beta * r[y, x])*deltat + sigma * np.sqrt(deltat) * \
                          np.random.normal(0, 1)
    plt.plot(t, r[y])

plt.title(
    'O-U-V \n Sims %d | Steps %d | Sigma %.2f | alpha %.2f | r0 %.2f' % (nsim, N, sigma, alpha, r0))
plt.xlabel('Steps')
plt.ylabel('Interest rate')
#plt.ylim(60,120)
plt.show()

#plt.plot([1,2,3,4,5,6,7,8,9])
#plt.show()
plt.savefig('O-U-V.png')

```
